## K-Means

```python
# import específico para meu computador, pode ignorá-lo
import warnings
warnings.filterwarnings('ignore')

#importando bibliotecas
import numpy  as np
import pandas as pd 

from sklearn    import cluster as ct
from sklearn    import metrics as mt
from matplotlib import pyplot as plt
```

```python
X = pd.read_csv('X_dataset.csv')
```

```python
X.shape
```

```python
n_cluster = np.arange(2,11)
ss_avglist =[]

for k in n_cluster:
    #definition
    kmeans = ct.KMeans( n_clusters = k,n_init=10)
    
    # fit
    labels = kmeans.fit_predict( X )
    
    # performance
    ss_avg = round(mt.silhouette_score( X, labels ),3)
    ss_avglist.append(ss_avg)
    
    print('Cluster K:{} ,SS:{}'.format(k , ss_avg))
```

**Para máxima SS, K é:**

```python
max_ss = max(ss_avglist)
k_max = ss_avglist.index(max_ss)

print ('Para K:{} , SS:{}'.format(n_cluster[k_max], max_ss))
```

## Gráfico Elbow Method

```python
plt.plot(n_cluster , ss_avglist , marker='o');
plt.ylabel ('Silhouette Score - SS');
plt.xlabel ('K value');
```
