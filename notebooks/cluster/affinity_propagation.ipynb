## Affinity Propagation

```python
#importando bibliotecas
import numpy  as np
import pandas as pd 

from sklearn    import cluster as ct
from sklearn    import metrics as mt
from matplotlib import pyplot as plt
```

```python
X = pd.read_csv('X_dataset.csv')
```

```python
p = np.arange(-1,-61,-1)
ss_list = []
for i in range(len(p)):
    #definition
    model = ct.AffinityPropagation (preference = p[i])
    
    #fit
    model.fit ( X )
    
    #predict
    labels = model.predict( X )

    #silhouette score
    ss = mt.silhouette_score( X , labels)
    ss_list.append( ss )

    #plot 
    plt.plot ( p[:i] , ss_list[:i])
    plt.title( 'Affinity Propagation - Convergence' )
    plt.xlabel ('Preference')
    plt.ylabel ('Silhouette Score - SS')
    
```

### Final Clustering

```python
ss_max = max( ss_list )
print ('Máximo SS:{}'.format(ss_max))
```
